```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
import pickle
def save_obj(obj, name):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
Q = load_obj("8_2-SMALL-irace")
df10_0 = pd.DataFrame.from_dict(Q)
df10_0  = df10_0.transpose()
df10_0  = df10_0.rename({'terminal' : (0,0)})
df10_0  = df10_0.sort_index()

%matplotlib inline
vals = df10_0.index.values
vals = sorted(vals, key=lambda x: math.atan2(x[1],x[0]))
df = df10_0.loc[vals]
A = df.as_matrix()
print(A.shape)
plt.figure(figsize=(10,10))
im = plt.imshow(A, cmap="gist_heat", interpolation='nearest',)
im.axes.set_aspect(A.shape[1] / A.shape[0])
```
